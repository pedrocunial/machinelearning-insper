# Árvores de decisão

Uma árvore de decisão é uma árvore binária onde os vários nós não-terminais correspondem a decisões acerca da amostra sendo analisada, e as folhas correspondem aos valores de predição. Por exemplo, considere o famoso dataset "Iris" e o problema de classificar as flores em "virginica", "versicolor" e "setosa".

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
y = pd.Series(iris['target'], name='class')
```

```python
def plot_hist(X, y):
    bins = np.linspace(X.min(), X.max(), 30)
    for c in y.value_counts().index:
        plt.hist(X[y == c].values, bins=bins, label=str(c), 
                 alpha=0.5, color=plt.cm.Set1.colors[c])

def plot_scatter(X, y):
    cols = X.columns
    n = len(cols)
    for i in range(n):
        plt.subplot(n, n, i*n + i + 1)
        plot_hist(X[cols[i]], y)
        plt.xlabel(cols[i])

        for j in range(i+1, n):
            plt.subplot(n, n, i*n + j + 1)
            plt.scatter(X[cols[j]], X[cols[i]], c=y, cmap=plt.cm.Set1, edgecolor='k')
    
plt.figure(figsize=(12, 12))
plot_scatter(X, y)
plt.show()
```

 Para simplificar nossa explanação, vamos nos limitar ao uso de apenas duas características: sepal width e petal width

```python
X_simple = X[['petal length (cm)', 'petal width (cm)']]
plt.figure(figsize=(8, 8))
plot_scatter(X_simple, y)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clf.fit(X_simple, y)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(tree_clf, out_file="iris_tree.dot", 
                feature_names=['petal length (cm)', 'petal width (cm)'],
                class_names=iris.target_names, rounded=True, filled=True)
```

**R:** 
 - O *Classification And Regression Tree* é um algoritmo para treinamento de árvores de decisão baseado na divisão do conjunto de treinamento em dois por um dado threshhold de um paramêtro (por exemplo, tamanho de pétala) e assim sucessivamente
 
 - Modelos paramétricos tem um número finito e determinado de parâmetros (por exemplo, uma regressão linear simples possui dois parâmetros, $\theta_0$ e $\theta_1$, enquanto os não-paramétricos utilizam um $\theta$ de "infinitas dimensões" e que só poderia ser descrito de maneira correta por uma função ou algo similar
 
 - O principal hiperparâmetro de uma árvore de decisão é o número mínimo de amostras em um nó-folha. Se não restringirmos este parâmetro, cada uma das amostras do conjunto de treinamento será mapeada em uma única folha (a menos que sejam EXATAMENTE o mesmo valor Y) o que, por definição, é um caso de overfitting
 
 - O *CART* retorna um valor absoluto para cada entrada. Em um modelo de classificação, isso significa um "grupo" no qual se encaixa, mas no caso do regressor, isso significa que para uma certa taxa de dados de entrada, o valor de saída será exatamente o mesmo, tal que uma reta diagonal provavelmente ficaria parecendo uma escada
 
 - Pois a árvore de decisão faz escolhas com base em diferenças em um mesmo parâmetro a partir de um valor (threshhold). Desta forma, não é relevante se os dados estão ou não normalizados ou re-escalados de qualquer forma
 
 - Aproximadamente 1000 horas, a complexidade para encontrar "a árvore perfeita" para um modelo é de $O(2^n)$, onde $n$ é o tamanho do dataset. Sendo o segundo dataset 10x maior do que o primeiro, esperá-se um aumento de 1024x

```python
## exercício 7

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

x_moons, y_moons = make_moons(n_samples=10000, noise=0.4)

x_train, x_test, y_train, y_test = train_test_split(x_moons, y_moons, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV

moon_clf = DecisionTreeClassifier(random_state=42)
params = {
    'min_samples_leaf': [10, 50, 100, 300, 500, 1000],
    'max_depth': [10, 20, 50, 100, 200]
}

gscv = GridSearchCV(moon_clf, params)
gscv.fit(x_train, y_train)
```

```python
gscv.best_params_
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_train, gscv.predict(x_train))
```

```python
accuracy_score(y_test, gscv.predict(x_test))
```

```python
## exercício 8 - grow a forest
from sklearn.model_selection import StratifiedShuffleSplit

sss = StratifiedShuffleSplit(n_splits=1000, test_size=0.1,
                             random_state=42)

x_forest = []
y_forest = []

for train_idx, test_idx in sss.split(x_moons, y_moons):
    x_forest.append((x_moons[train_idx], x_moons[test_idx]))
    y_forest.append((y_moons[train_idx], y_moons[test_idx]))
    
```

```python
forest_clf = [DecisionTreeClassifier(random_state=42,
                                    min_samples_leaf=100,
                                    max_depth=10) \
              for _ in range(len(x_forest))]

for f_c, x, y in zip(forest_clf, x_forest, y_forest):
    f_c.fit(x[0], y[0])
```
