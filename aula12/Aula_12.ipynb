## Teorema de Bayes

Considere uma amostra de treinamento $(\mathbf{x}^{(i)}, y^{(i)})$ em um problema de classificação. Esta amostra foi obtida ao acaso, a partir da população deste problema. Podemos dizer que a amostra de treinamento foi obtida de uma variável aleatória (ou melhor, de um vetor aleatório). 

Podemos então pensar na probabilidade conjunta dos valores de *features* e da classe da amostra:

$$P(\mathbf{x}, y) = \text{Probabilidade de se observar caracteristicas $\mathbf{x}$ e classe $y$ simultâneamente}$$

Por exemplo, no caso dos jogadores de basquete, se $\mathbf{x} = (\text{altura}, \text{peso})$ e $y = \text{Posição}$, então 
- para $\mathbf{x} = (2.10, 125)$ e $y=\text{'Pivô'}$ temos que $P\left(\mathbf{x}, y\right)$ é alta,
- para $\mathbf{x} = (1.60, 75)$ e $y=\text{'Pivô'}$ temos que $P\left(\mathbf{x}, y\right)$ é baixa,
- para $\mathbf{x} = (2.10, 125)$ e $y=\text{'Armador'}$ temos que $P\left(\mathbf{x}, y\right)$ é baixa.

Pelo teorema de Bayes:

$$P(\mathbf{x}, y) = P(\mathbf{x} | y) P(y) = P(y | \mathbf{x}) P(\mathbf{x})$$

portanto

$$P(y | \mathbf{x}) = \frac{P(\mathbf{x} | y) P(y)}{P(\mathbf{x})}$$

O termo $P(\mathbf{x} | y)$ é chamado de **verossimilhança** (em inglês: *likelihood*), e corresponde à distribuição probabilística dos valores das *features* dos elementos de uma dada classe. Por exemplo: $P(\mathbf{x} \,| \, \text{'Pivô'})$ é a probabilidade das alturas e pesos de pivôs de basquete, enquanto $P(\mathbf{x} \, | \, \text{'Armador'})$ é a probabilidade de alturas e pesos de armadores, provavelmente formando uma concentração de pontos mais abaixo dos pivôs em ambas as dimensões.

O termo $P(y)$ é chamado de probabilidade **à priori** (em inglês: *prior*). Representa a probabilidade de que uma amostra escolhida ao acaso pertença à classe $y$ (sem ao menos olhar para as *features* $\mathbf{x}$). Por exemplo, $P('Pivô')$ é a probabilidade de que um jogador de basquete escolhido ao acaso seja um pivô.

O termo $P(\mathbf{x})$ é a probabilidade à priori das características (juntando todas as classes). Costuma ser meio irrelevante para problemas de classificação, por razões que ficarão evidentes em breve. E é um termo redundante de qualquer forma:

$$P(\mathbf{x}) = 
\sum_{i = 1}^{N_{classes}} P(\mathbf{x}, y = C_i) = \sum_{i = 1}^{N_{classes}} P(\mathbf{x} \, | \, y = C_i) P(y = C_i)$$

e portanto $P(\mathbf{x})$ pode ser obtido a partir do *prior* e do *likelihood*.

O termo $P(y \, | \, \mathbf{x})$ é chamado de probabilidade **à posteriori** (em inglês: *posterior*), e representa a probabilidade de que a amostra pertença à classe $y$ após observarmos suas características $\mathbf{x}$. Compare com a *prior* $P(y)$. Por exemplo: $P(\text{'Pivô'})$ deve ser 20% (pois todo time de cinco tem um pivô) mas $P(\text{'Pivô'} \, | \, (\text{altura} = 2.20 \text{m}, \text{peso} = 135 \text{kg}))$ deve ser bem mais alta!

## Bayes e classificação

Suponha que queremos saber qual a classe mais provável de uma amostra baseado em suas características. Ou seja, queremos comparar $P(y = C_i \, | \, \mathbf{x})$ e $P(y = C_j \, | \, \mathbf{x})$ para determinar se, baseado nas características $\mathbf{x}$ que observamos em um objeto, se é mais provável que ele pertença á classe $C_i$ ou $C_j$. Aplicando o teorema de Bayes, temos:

$$P(y = C_i \, | \, \mathbf{x}) > P(y = C_j \, | \, \mathbf{x})
\Leftrightarrow \\
\frac{P(\mathbf{x} \, | \, y = C_i) P(y = C_i)}{P(\mathbf{x})} > 
\frac{P(\mathbf{x} \, | \, y = C_j) P(y = C_j)}{P(\mathbf{x})} \Leftrightarrow \\
P(\mathbf{x} \, | \, y = C_i) P(y = C_i) > P(\mathbf{x} \, | \, y = C_j) P(y = C_j)$$

Logo, se tivermos o *prior* e o *likelihood* de cada classe do nosso problema, conseguimos determinar a classe mais provável para nossas novas amostras através da comparação dos produtos *prior* $\times$ *likelihood* de cada classe.

Em geral, se desejamos maximizar a acurácia:

$$C = \text{argmax}_{C_i}\left\{P(\mathbf{x} \, | \, y = C_i) P(y = C_i) \right\}$$

**R:** 
 - A distância de Mahalanobis é uma medida utilizada para a detetecção de similaridade entre amostras. Ela trabalha com base em correlações entre variáveis. Ao contrário da distância de Euclidiana a distância de Mahalanobis independe de escala.
 - O classificador K-NN funciona assumindo que a classe a qual uma entrada pertence é a mesma dos seus *k* vizinhos mais proximos, sendo *k* um número inteiro e a distância entre o ponto e seus vizinhos sendo calculada pelo método de Mahalanobis ou pela distância Euclidiana. Por ter que calcular os *k* nós mais próximos para cada predição, o classificador apresenta uma enorme perda de performance para entradas largas.

```python
def cov_matrix(s1, s2, t):
    rad = t * np.pi / 180.0
    S = np.array([[np.sqrt(s1)*np.cos(rad), np.sqrt(s2)*np.sin(rad)], [-np.sqrt(s1)*np.sin(rad), np.sqrt(s2)*np.cos(rad)]])
    return S.T.dot(S)

def make_dataset(mx, my, sx, sy, ang, n):
    X = np.random.multivariate_normal([mx, my], cov_matrix(sx, sy, ang), n)
    y = np.zeros((n, 1))
    return X, y
    
X1, y1 = make_dataset(0, 0, 10, 1, 30, 1000)
X2, y2 = make_dataset(5, 5, 10, 1, 30, 1000)
plt.figure(figsize=(8,6))
plt.scatter(X1[:,0], X1[:,1], c='b', marker='o', alpha=0.1)
plt.scatter(X2[:,0], X2[:,1], c='r', marker='o', alpha=0.1)
plt.axis('equal')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

# print(X1, '\n\n', y1)
y1_map = np.vectorize(lambda x: 1)(y1)
y2_map = np.vectorize(lambda x: 2)(y2)

y_final = np.concatenate((y1_map, y2_map))
X_final = np.concatenate((X1, X2))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_final,
    y_final,
    test_size=.33,
    random_state=RANDOM_SEED
)

clf = KNeighborsClassifier(metric='euclidean', weights='distance')
```

```python
from sklearn.model_selection import GridSearchCV

# teoricamente, o melhor desempenho é com n=1, mas eu me recuso
# a aceitar que isso possa estar ou ser correto
gs = GridSearchCV(clf, {
    'n_neighbors': [3, 5, 10]
})

gs.fit(X_train, y_train.ravel())
```

```python
gs.best_params_
```

```python
from sklearn.metrics import accuracy_score

y_pred = gs.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
maha_clf = KNeighborsClassifier(
    metric='mahalanobis',
    metric_params={'V': np.cov(X_train, rowvar=False)},
    weights='distance'
)

maha_gs = GridSearchCV(maha_clf, {
    'n_neighbors': [3, 5, 10],
})

maha_gs.fit(X_train, y_train.ravel())
```
